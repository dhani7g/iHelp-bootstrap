```python
# To decrease the total vocabulary size and enhance the model’s capacity to generalize to new data, all texts are converted to lowercase. We then expand the contradictions to improve the consistency of the text among different sections. URLs, punctuation, stop words and white spaces are removed and accented characters are replaced with standard characters. give code

# Convert all text to lowercase
df = df.apply(lambda x: x.astype(str).str.lower())

# Expand contradictions
df = df.applymap(lambda x: ' '.join(x.split()) if isinstance(x, str) else x)

# Remove URLs
df = df.applymap(lambda x: re.sub(r'https?://\S+', '', x) if isinstance(x, str) else x)

# Remove punctuation
df = df.applymap(lambda x: re.sub(r'[^\w\s]', ' ', x) if isinstance(x, str) else x)

# Remove stop words
stop_words = set(stopwords.words('english'))
df = df.applymap(lambda x: ' '.join([word for word in x.split() if word not in stop_words]) if isinstance(x, str) else x)

# Remove white spaces
df = df.applymap(lambda x: ' '.join(x.split()) if isinstance(x, str) else x)

# Replace accented characters with standard characters
df = df.applymap(lambda x: unidecode(x) if isinstance(x, str) else x)

# Display the modified DataFrame
print(df.head())
```

```python
from collections import defaultdict

def tokenize_and_map_to_int(text_column):
    token_to_int = defaultdict(lambda: len(token_to_int))
    token_ids_list = []
    token_to_int_list = []
    for text in text_column:
        token_to_int_current = {}
        if isinstance(text, str):
            tokens = text.split()  # Split text into tokens (words)
            token_ids = [token_to_int[token] for token in tokens]  # Map tokens to integers
            token_ids_list.append(token_ids)
            # Create a dictionary mapping tokens to their corresponding integer IDs
            token_to_int_current = {token: token_ids[idx] for idx, token in enumerate(tokens)}
        else:
            token_ids_list.append([])
        token_to_int_list.append(token_to_int_current)
    return token_ids_list, token_to_int_list

# Define the columns containing text data
text_columns = ['candidate_name', 'description', 'location', 'party', 'position', 'state', 'url', 'gender', 'twitter handle', 'twitter name']

# Tokenize and map tokens to integers for each text column
token_ids_dict = {}
token_to_int_dict = {}
for column in text_columns:
    token_ids_dict[column], token_to_int_dict[column] = tokenize_and_map_to_int(df[column])

# Print token IDs and token-to-integer mappings for each text column
for column in text_columns:
    print(f"Token IDs for '{column}': {token_ids_dict[column]}")
    print(f"Token to Integer Mapping for '{column}': {token_to_int_dict[column]}")
    print()
```

```python
# Define function to label accounts based on description
def label_spam(description):
    # Define keywords indicating spam accounts
    spam_keywords = ['free', 'discount', 'offer', 'promotion', 'sale', 'deal', 'win', 'prize', 'earn']

    # Check if any spam keyword is present in the description
    if any(keyword in description.lower() for keyword in spam_keywords):
        return 1  # Spam
    else:
        return 0  # Genuine

# Apply function to label each account
df['target'] = df['description'].apply(label_spam)

# Print distribution of labels
print(df['target'].value_counts())

# Save labeled dataset
df.to_csv("labeled_dataset.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Handle candidate name (categorical)
candidate_encoder = OneHotEncoder()
candidate_encoded = candidate_encoder.fit_transform(df[['candidate_name']])

# Handle Twitter name (text)
tfidf_vectorizer = TfidfVectorizer()
twitter_name_encoded = tfidf_vectorizer.fit_transform(df['twitter name'])

# Combine encoded features with other numeric features
X_numeric = df[['followers_count', 'friends_count', 'co_occurring_word_distribution']]  # Numeric features
X_encoded = np.hstack((candidate_encoded.toarray(), twitter_name_encoded.toarray(), X_numeric))

# Split the data into features (X) and target variable (y)
y = df['target']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train the XGBoost model
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=15, learning_rate=0.1)  # Use the optimal parameters
xgb_model.fit(X_train, y_train)

# Evaluate the model
y_pred = xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Define models
svm_poly_model = SVC(kernel='poly', random_state=42)
svm_rbf_model = SVC(kernel='rbf', random_state=42)
lr_model = LogisticRegression(random_state=42)
nn_model = KNeighborsClassifier(n_neighbors=3)
gbm_model = GradientBoostingClassifier(random_state=42)

# Train models
svm_poly_model.fit(X_train, y_train)
svm_rbf_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
nn_model.fit(X_train, y_train)
gbm_model.fit(X_train, y_train)

# Evaluate models
svm_poly_accuracy = svm_poly_model.score(X_test, y_test)
svm_rbf_accuracy = svm_rbf_model.score(X_test, y_test)
lr_accuracy = lr_model.score(X_test, y_test)
nn_accuracy = nn_model.score(X_test, y_test)
gbm_accuracy = gbm_model.score(X_test, y_test)

print("SVM Poly Accuracy:", svm_poly_accuracy)
print("SVM RBF Accuracy:", svm_rbf_accuracy)
print("Logistic Regression Accuracy:", lr_accuracy)
print("3NN Accuracy:", nn_accuracy)
print("Gradient Boosting Accuracy:", gbm_accuracy)
```
